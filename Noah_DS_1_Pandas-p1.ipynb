### 2.3. Describe text data

* When we use `.describe()`, by default it is executed with a parameter `include = None`, which mean that the function is applied on numeric values only.
* We can change this parameter to `include = ['object']` so that the function will be applied on object data, and with a different set of described parameters.

```python
car_sales.describe(include = ['object'])
```

### 3.1. Selecting a column

```python
# Selecting a column
car_sales['Make']
```

* The drawback of the 2nd way is that we cannot do that for column names with spaces.
* E.g. for the column 'Odometer (KM)', we cannot do `car_sales.Odometer (KM)` but has to do `car_sales['Odometer (KM)']`.

```python
# We can also assign a variable to a list of target column and access it
# This is very useful in some context

target_col = ['Price']
car_sales[target_col]
```

```python
car_sales.Colour.values

# see a column as an array/series
```

### 3.2. Selecting multiple columns

```python
# Method 1. Instead of providing a string for the col name, here we need to provide a nested list of the col names
# the output will be in the DataFrame format

car_sales[['Price', 'Make']]
```

```python
# Method 2: assigning the target column as a list variable

target_col_2 = ['Price', 'Make']
car_sales[target_col_2]
```

* Note that if we want a DataFrame structure for a Series, then we need to use the nested list as above.
* This is because there are functions that are applicable on DataFrame structure only.

```python
print(car_sales['Make'].shape) # 1 dimension

print(car_sales[['Make']].shape) # 2 dimensions, i.e. row x column
```

### 3.3. `.unique` and `.nunique`

```python
# Let's try another example
NFLX_title = pd.read_csv('data/netflix_titles.csv', sep = "|", index_col = 0, header = 0)
NFLX_title.head()
```

```python
NFLX_title.title.unique()

# this will list out the unique element in the column
```

```python
NFLX_title.title.nunique()

# this will count the number of unique element in the column
```

```python
NFLX_title.title.nunique() == len(NFLX_title.title.unique())

# gut check
```

### 3.4. `.nlargest` and `.nsmallest`

```python
NFLX_title['release_year'].nlargest(5)

# top largest values in the column
```

```python
NFLX_title['release_year'].nsmallest(3)

# top smallest values in the column
```

```python
# We can also display the whole row corresponding to the nlargest/nsmallest entries of a Series
# For this syntax, we need to specify the conditional column 

NFLX_title.nsmallest(3, ['release_year'])
```

### 3.5. `.value_counts`

```python
NFLX_title['country'].value_counts()

# count the frequency of each element in the column
```

```python
# there are useful parameters for it to explore

NFLX_title['country'].value_counts(ascending = True)
```

```python
# A powerful example of this function is that we can count a combination of data
# note that we need to use the DataFrame structure

NFLX_title[['country', 'release_year']].value_counts()
```
