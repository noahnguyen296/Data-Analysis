### 3.4. `.sort_values`

* Important parameters for `.sort_values()` are:
- `by = 'conditional column'`;
- `ascending = True/False`;
- `axis = 0/1`;
- `inplace = True/False`.

```python
# Another dataset example
houses = pd.read_csv('data/kc_house_data.csv')
houses.head()
```

```python
houses.sort_values(by = 'price')

# by default ascending = True
```

### 3.5. Sort by multiple columns

```python
houses.sort_values(by = ['bedrooms', 'bathrooms'])

# Sort by bedrooms, then sort by bathrooms; both are in ascending order by default
```

```python
houses.sort_values(by = ['bedrooms', 'bathrooms'], ascending = [True, False])

# Sort by bedrooms in ascending, then sort by bathrooms in descending
```

### 3.6. Sort string data

* `.sort_values()` are applicable to string, with the ascending order as ABC...XYZabc...xyz.
* A noteworthy parameter is `key` which provides a way to convert the string data into different format directly.

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head(2)
```

```python
titanic.sort_values(by = 'name')
```

```python
# Now we want to convert all the 'name' to lower case as we prefer a->z
# we can access the string function and use .lower() on the column first, then sort; otherwise we have a direct parameter

titanic.sort_values(by = 'name', key = lambda column: column.str.lower())
```

* Now we see the clear advantage of using `key`: it does not modify the string permanently, rather it applies the function `.lower()` as an interim step for the purpose of sorting only.

### 3.10. `.loc` and `.iloc` with Series

* A simple application of `.iloc.` and `.loc` is for row selection.
* But the functions also allow us to add arguments for columns selection:
1) .loc[rows arguments, columns arguments]
2) .iloc[rows arguments, columns arguments]
* The arguments for the rows and columns are similar, as demonstrated above; they only differ in the axis applied.

```python
# Example 1
(
    happiness.loc[
    'Togo':'Turkey', # row arguments
    ['upperwhisker', 'lowerwhisker', 'Generosity'] # column arguments
    ]
)
# TIP: for complex arguments, we can write the code similar to this so that we can fix easier
```

```python
# Example 2
(
    happiness.loc[
    ['Denmark', 'Sweden', 'Norway'], # row arguments
    'Explained by: Log GDP per capita':'Explained by: Perceptions of corruption' # column arguments
    ]
) 
# TIP: for complex arguments, we can write the code similar to this so that we can fix easier
```

```python
# Example 3
(
    happiness.iloc[
    2:27:3, # row arguments
    1:10:2 # column arguments
    ]
) 
# TIP: for complex arguments, we can write the code similar to this so that we can fix easier
```

```python
# Example 4
(
    happiness.iloc[
    -10:, # row arguments (final 10 countries)
    : # column arguments (all columns)
    ]
) 
# TIP: for complex arguments, we can write the code similar to this so that we can fix easier
```

### 4.1. Boolean indexing with comparison operators

```python
# Filtering data = filter data given conditions
# An essential datatype for filtering data is Boolean data, i.e. True/False

print(3 > 5)
print(1 != 0)
print('a' == 'a')
```

```python
# If we apply the argument for a Series, it will return a Series of the same but with Boolean data
titanic.sex == 'female'
```

```python
# Using the Boolean data to filter data, we will receive data corresponding to True

titanic[titanic.sex == 'female'] # filtering rows with sex == 'female'
```

```python
# another example
titanic[titanic.pclass == 3] # filtering rows for people at pclass 3
```

```python
# another example (!= means NOT)

titanic[titanic.embarked != 'S'] # filtering rows for people at pclass different than 3
```

```python
# another example with other data

houses[houses.price > 5000000] # selecting houses with price > 5,000,000
```

### 4.2. `.between`

```python
houses[houses.bedrooms.between(2, 4)]

# optional parameters: inclusive = 'both' / 'neither' / 'left' / 'right'
```

### 4.3. `.isin`

```python
# We can employ .isin() to filter data given a conditional list

# E.g. Filtering by specific tickets
target_ticket_list = ['PC 17482', 'S.O.P. 1166', 'C.A. 31030', 'W./C. 14263']
titanic[titanic['ticket'].isin(target_ticket_list)]
```

```python
# more complex example

(
    happiness[
    happiness['Regional indicator'].isin(['Southeast Asia', 'Sub-Saharan Africa'])
    ]                                      # select 'Regional indicator' in 'Southeast Asia' and 'Sub-Saharan Africa'
)[['Logged GDP per capita', 'Generosity']] # then further select these 2 columns to look at
```

### 4.4. Combining conditions using `AND`

```python
# The syntax for AND is: &

female = titanic.sex == 'female'  # first Boolean series
died = titanic.survived == 0       # second Boolean series

female & died # combined Boolean series satisfying both conditions
```

```python
# now we can apply the combined condition
# this gives us rows corresponding to females that has died

titanic[female & died]
```

```python
houses
```

```python
# try another example with a direct code line

(
    houses[
    (houses.price <= 2000000) &        # 1st condition based on price
    (houses.bedrooms.between(2, 3)) &  # 2nd condition based on number of bedrooms
    (houses.waterfront == 0)           # 3rd condition based on waterfront
    ]
)
```

* Note that for direct coding, we need to wrap each condition in the parentheses, and match each condition with `&`

### 4.5. Combining conditions using `OR`

* The syntax for `OR` is similar to `AND`, but now we use `|`

```python
# first simple example

titanic[(titanic.pclass == 2) | (titanic.pclass == 3)]
```

* We can see that the application of OR is similar to `.isin()` to some extent.
* E.g. `(titanic.pclass == 2) | (titanic.pclass == 3)` is equivalent to `titanic.pclass.isin([2, 3])`.

```python
# More complex example: using both AND and OR
embarked_condtion = ['C', 'S']

(titanic[
 ((titanic['pclass'] == 2) | (titanic['pclass'] == 3) ) &  # 1st condition using OR
 (titanic['survived'] == 0) &                              # 2nd condition based on survived
 (titanic['embarked'].isin(embarked_condtion))             # 3rd contion based on embarked
]                                                          # then combine using AND; remember to wrap each condition in ()
)
```

### 4.7. `.isna` and `.notna`

```python
NFLX_titles = pd.read_csv('data/netflix_titles.csv', sep = "|", index_col = 0, header = 0)
NFLX_titles.info()
```

* Using `.info()`, we see that there are 8807 rows. However, some columns contain fewer non-null than that, implying the existence of NaN entries.
* `.isna` and `.notna` are the methods used to filter NaN and non-Nan values.

```python
(
    NFLX_titles[
    (NFLX_titles.director.isna()) & 
    (NFLX_titles.cast.isna())
    ] 
)
# filtering the rows with missing values on the columns 'director' AND 'cast'
```

### 4.8. Filtering + plot

```python
# We will have a more detail practice on plotting
# Here we apply the filtering practice so far with simple plotting
# using the predefined condition 'female' above

(
    titanic[female]['survived']
    .value_counts()
    .plot(kind = 'pie')
);  # Sort the female data, count the numbers of survivor/non-survivor and plot them as a pie chart
```

```python
# Application of bitwise negation (~)

(
    titanic[~female]['survived']
    .value_counts()
    .plot(kind = 'pie')
);  # Sort the male data, count the numbers of survivor/non-survivor and plot them as a pie chart
```
