### 8.2. Multiple plots - figure

```python
a = np.arange(6)
a2 = a**2
a3 = a**3
```

```python
# If we plot multiple charts in a same code cell, then they will be into a single chart

plt.plot(a, a)
plt.plot(a, a2)
plt.plot(a, a3);
```

Here we have some essential concepts:
* A **figure** is an object that contain multiple separate graphs. We can think of it as the largest canvas that contains small items (the graphs).
* A figure can be divided into multiple rows and columns, with each entries corresponding to a single graph.
* Each single graph in a figure is termed an **axes**.

```python
# Let's do an example of separating the graphs above

plt.figure()
plt.plot(a, a)

plt.figure()
plt.plot(a, a2)

plt.figure()
plt.plot(a, a3);
```

* To separate the graphs, we need to call `plt.figure()` before plotting each of them.

### 8.5. Adjusting line aesthetics

Some useful parameters to control the visualization of the plots:
1. `color` (or `c`). It allows for English words such as 'red', 'navy', 'darkblue', 'olive', and hex codes such as '#ff6b6b' too.
2. `linewidth`. This is to control how thick the line is.
3. `linestyle`. This is to control to type of the line, e.g. 'dashed' (or simply '--'), '-.' (dashed then dot) etc.
4. `marker`. This is to add markers on the data points, e.g. 'o' (circle), '^' (triangle up), 'v' (triangle down) etc.
5. `markersize`. This is to control the size of the markers.
6. `markerfacecolor`. This is to control the color of the markers.

More detail: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# example

plt.figure(figsize = (13,4))
plt.plot(a, a3,
    color = 'navy',
    linewidth = 1.5,
    linestyle = '-.',
    marker = '*',
    markersize = 10,
    markerfacecolor = 'orange'
        );
```

### 8.6. Plot labels and titles

* We can call `plt.xlabel()` and `plt.ylabel()` to label the x-axis and y-axis.
* We can use the parameter `title` to create a name for the graph/figure. We can also use `plt.title` as a separate line.
* Besides the main arguments of the label/title names, we have `loc` to control the position, `fontsize` to control the text size, `color` to control the color of the text.

```python
# example

plt.figure(figsize = (13,4))
plt.plot(a, a3,
    color = 'navy',
    linewidth = 1.5,
    linestyle = '-.',
    marker = '*',
    markersize = 10,
    markerfacecolor = 'orange'
        )

plt.title('X vs X-cubed', loc = 'center', color = 'darkred', fontsize = 14)
plt.xlabel('X', fontsize = 12, color = 'navy')
plt.ylabel('X-cubed', fontsize = 12, color = 'purple');
```

### 8.7. Changing x- and y-ticks

* `plt.xticks()` and `plt.yticks` are methods used to control the display of x- and y-values.
* Noteworthy parameters include `ticks` (an array of values we want to display), `labels` (a list of label for the values), `rotation` (adjusting the direction of the values display).

```python
plt.figure(figsize = (13,4))
BTC.Close.plot()

plt.yticks([5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000], 
           labels = ['5k', '10k', '20k', '30k', '40k', '50k', '60k', '70k'])

plt.xticks(rotation = 45)
;
```

### 8.8. Add legends and grids

* We can call `plt.legend()` to add legends.
* Noteworthy parameters include
1. `loc` to specify the location of the legend such as 'left', 'center', 'right', 'upper left', 'lower right', 'upper center','best' ...,;
2. `fontsize` to specify the size of the legends;
3. `labelcolor` to control to color of the text in the legend;
4. `shadow = True/False` to add shadow;
5. `fancybox = True/False`.
6. `facecolor` to control the color of the legend box

* We can call `plt.grid(True)` to add the gridlines to the graph. One useful parameter is, says, `alpha = 0.3` to control the opacity.

```python
plt.figure(figsize = (13,4))
BTC.Close.plot()

plt.legend(
    loc = 'best',
    fontsize = 12,
    labelcolor = 'darkred',
    shadow = True,
    fancybox = True,
    facecolor = 'yellow'
)

plt.grid(True, alpha = 0.4);
```

### 8.9 Bar charts

```python
plants = ['Morning glory', 'Spinach', 'Broccoli', 'Tomato', 'Carrot', 'Cucumber']
died = [12, 13, 5, 21, 7, 4]
germinated = [55, 43, 57, 66, 44, 77]
```

```python
# To call a bar chart, we use plt.bar()
# Example 1

plt.figure(figsize = (8, 4))
plt.bar(
    plants,               # x
    germinated,           # y
    color = 'royalblue',  # control the bar color
    width = 0.5           # control the bar width
)

plt.grid(alpha = 0.3);
```

```python
# Example 2: multiple bar charts (double bars chart)
# Basically we will adjust by the parameters 'width' and 'align'

plt.figure(figsize = (8, 4))

# 1st bar chart
plt.bar(
    plants,               # x
    died,                 # y
    color = 'darkred',    # control the bar color
    width = 0.4,          # control the bar width
    align = 'edge',
    label =  'Died'
)

# 2nd bar chart
plt.bar(
    plants,               # x
    germinated,           # y
    color = 'royalblue',  # control the bar color
    width = 0.3,
    label = 'Germinated'
)

plt.grid(alpha = 0.3)
plt.legend();
```

```python
# Example 3: stacked bar chart --> control by 'bottom'

plt.figure(figsize = (8, 4))

# 1st bar chart
plt.bar(
    plants,
    died,
    color = 'darkred',
    bottom = germinated, # specify which data is displayed as the bottom bar chart
    label = 'Died'
)

# 2nd bar chart
plt.bar(
    plants,
    germinated,
    color = 'royalblue',
    label = 'Germinated'
)

plt.grid(alpha = 0.3)
plt.legend(shadow = True, fancybox = True, facecolor = 'white');
```

```python
# Example 4: to plot horizontal bar chart, we call plt.barh
# Note that for the stacked bar, now we change 'bottom' to 'left'

plt.figure(figsize = (8, 4))

# 1st bar chart
plt.barh(
    plants,
    died,
    color = 'darkred',
    left = germinated, # specify which data is displayed as the bottom bar chart
    label = 'Died'
)

# 2nd bar chart
plt.barh(
    plants,
    germinated,
    color = 'royalblue',
    label = 'Germinated'
)

plt.grid(alpha = 0.3)
plt.legend(shadow = True, fancybox = True, facecolor = 'white');
```

### 8.13. Subplots

* In this part, we demonstrate of how to control the position of each graph in a large figure of multiple graphs.
* It is a little confusing but some tips are:
1. We still call `plt.figure()` to generate a whole figure. The parameters for it are applied to all graphs.
2. Before plotting each graph, we call `plot.subplot()` with the input is of the structure (number of rows, number of columns, the position of the graph). It would be useful to consider the figure as a matrix here. The position of the graph is counted from left to right and top to bottom.
3. Each graph can be adjusted with different methods and parameters.
4. To add a title for the whole figure, we call `plt.suptitle()`. It works similar as `plt.title()`.
6. For complex graphs, objects can be overlapped. We can call `plt.tight_layout()` and Matplotlib will attempt to organize the objects to improve the display.

```python
# Using the previous data
print(a)
print(a2)
print(a3)
```

```python
# Example 1: consider the figure as a matrix of 1 row and 3 columns to plot 3 separate graphs

plt.figure(figsize = (10, 4), dpi = 100) # this is applied to all of the following sub graphs
plt.suptitle('Example 1', fontsize = 12, color = 'black')

plt.subplot(1, 3, 1)               # 1 row, 3 columns, position 1
plt.plot(a, a, color = 'navy')
plt.title('X', fontsize = '11')

plt.subplot(1, 3, 2)               # 1 row, 3 columns, position 2
plt.plot(a, a2, color = 'darkred')
plt.title('X squared', fontsize = 14, color = 'darkred')

plt.subplot(1, 3, 3)               # 1 row, 3 columns, position 3
plt.plot(a, a3, color = 'darkgreen')
plt.title('X cubed')

plt.grid(alpha = 0.3) # we will see that this apply to only the final graph; hence we need to call it for each graph if needed
plt.tight_layout()
```

```python
# Example 2: vertical plots

plt.figure(figsize = (7, 7), dpi = 100) # in this case we may need to adjust the figsize
plt.suptitle('Example 1', fontsize = 12, color = 'black')

plt.subplot(3, 1, 1)               # 3 row, 1 columns, position 1
plt.plot(a, a, color = 'navy')
plt.title('X', fontsize = '11')

plt.subplot(3, 1, 2)               # 3 row, 1 columns, position 2
plt.plot(a, a2, color = 'darkred')
plt.title('X squared', fontsize = 14, color = 'darkred')

plt.subplot(3, 1, 3)               # 3 row, 1 columns, position 3
plt.plot(a, a3, color = 'darkgreen')
plt.title('X cubed')

plt.tight_layout()
```

```python
# Example 3: demonstrating how the graph position works

plt.figure(figsize = (7, 7), dpi = 100)
plt.suptitle('Example 1', fontsize = 12, color = 'black')

plt.subplot(2, 2, 1)               # 2 row, 2 columns, position 1
plt.plot(a, a, color = 'navy')
plt.title('X', fontsize = '11')

plt.subplot(2, 2, 2)               # 2 row, 2 columns, position 2
plt.plot(a, a2, color = 'darkred')
plt.title('X squared', fontsize = 14, color = 'darkred')

plt.subplot(2, 2, 3)               # 2 row, 2 columns, position 3
plt.plot(a, a3, color = 'darkgreen')
plt.title('X cubed')

plt.tight_layout()
```

* We can see that that graph 'X-cubed' is still in position 3. This shows how graph positions are applied: from left to right, from top to bottom.
